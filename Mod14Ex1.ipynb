gráficos referentes a base de dados de março, abril, maio, junho e dezembro
(numa célula só ou em várias).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)


```

```python
meses = ['03', '04', '05', '06', '12']
dfs = []

for mes in meses:
    df = pd.read_csv('SINASC_RO_2019.csv')
    df['MES'] = mes 
    dfs.append(df)

dados = pd.concat(dfs)
dados.head()


```

```python
nascimentos_mes = dados['MES'].value_counts().sort_index()

nascimentos_mes.plot(kind='bar', color='skyblue')
plt.title('Número de Nascimentos por Mês (Março, Abril, Maio, Junho, Dezembro)')
plt.xlabel('Mês')
plt.ylabel('Total de Nascimentos')
plt.xticks(rotation=0)
plt.show()



```

```python
nascimentos_mes = dados['MES'].value_counts().sort_index()

nascimentos_mes.plot(kind='bar', color='skyblue')
plt.title('Número de Nascimentos por Mês (Março, Abril, Maio, Junho, Dezembro)')
plt.xlabel('Mês')
plt.ylabel('Total de Nascimentos')
plt.xticks(rotation=0)
plt.show()

```

```python
sns.histplot(data=dados, x='IDADEMAE', bins=20, kde=True, color='orange')
plt.title('Distribuição da Idade das Mães')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()

```

```python
sns.countplot(data=dados, x='MES', hue='QTDFILVIVO', palette='Set2')
plt.title('Distribuição do Quantidade Filhos Vivos')
plt.xlabel('Mês')
plt.ylabel('Contagem')
plt.legend(title='QTDFILVIVO')
plt.show()
```

```python
sns.boxplot(data=dados, x='MES', y='PESO', palette='pastel')
plt.title('Distribuição do Peso ao Nascer por Mês')
plt.xlabel('Mês')
plt.ylabel('Peso (g)')
plt.show()
```

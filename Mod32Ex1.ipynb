# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t
import matplotlib.pyplot as plt
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
mu = 15     
sigma = 0.1  

x = 15.2

prob = 1 - norm.cdf(x, loc=mu, scale=sigma)

print(f"A probabilidade de um pacote pesar mais que {x} kg é {prob:.4f} (~{prob*100:.2f}%)")
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
mu = 15
sigma = 0.1
p = 0.95 

x_95 = norm.ppf(p, loc=mu, scale=sigma)

print(f"O percentil de ordem 95% para o peso é {x_95:.4f} kg")
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu código começa aqui
prob_norm = norm.cdf(Y, loc=0, scale=1)

prob_t = t.cdf(Y, df=5)

print("Y:", Y[:5])
print("P(Y1 <= y) - Normal:", prob_norm[:5])
print("P(Y2 <= y) - t-Student:", prob_t[:5])

```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
plt.plot(Y, prob_norm, label="Normal Padrão")
plt.plot(Y, prob_t, label="t-Student (df=5)")
plt.xlabel("Valores de Y")
plt.ylabel("P(Y <= y)")
plt.title("Função de Distribuição Acumulada (CDF)")
plt.legend()
plt.grid(True)
plt.show()
```

Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu código começa aqui
P = np.linspace(0, 1, 20)

quantis_norm = norm.ppf(P, loc=0, scale=1)

quantis_t = t.ppf(P, df=5)

print("P (probabilidades):")
print(P)
print("\nQuantis - Normal Padrão:")
print(quantis_norm)
print("\nQuantis - t-Student (df=5):")
print(quantis_t)
```

```python
plt.figure(figsize=(8,5))
plt.plot(P, quantis_norm, marker='o', label='Normal Padrão')
plt.plot(P, quantis_t, marker='x', label='t-Student (df=5)')
plt.xlabel('Probabilidade (P)')
plt.ylabel('Quantil (percentil)')
plt.title('Quantis da Normal Padrão e t-Student (df=5)')
plt.legend()
plt.grid(True)
plt.show()
```

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('previsao_de_renda.csv')

print(df.head())
````

````python
print(df['posse_de_imovel'].value_counts())
print(df['posse_de_veiculo'].value_counts())
````

````python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='posse_de_imovel', y='renda', data=df)
plt.title('Renda por Posse de Imóvel')
plt.xlabel('Possui Imóvel?')
plt.ylabel('Renda')

plt.subplot(1, 2, 2)
sns.boxplot(x='posse_de_veiculo', y='renda', data=df)
plt.title('Renda por Posse de Veículo')
plt.xlabel('Possui Veículo?')
plt.ylabel('Renda')

plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(6, 4))
df.groupby('posse_de_imovel')['renda'].mean().plot(kind='bar', color='teal')
plt.title('Renda Média por Posse de Imóvel')
plt.xlabel('Possui Imóvel?')
plt.ylabel('Renda Média')
plt.xticks(ticks=[0, 1], labels=['Não', 'Sim'], rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(6, 4))
df.groupby('posse_de_veiculo')['renda'].mean().plot(kind='bar', color='orange')
plt.title('Renda Média por Posse de Veículo')
plt.xlabel('Possui Veículo?')
plt.ylabel('Renda Média')
plt.xticks(ticks=[0, 1], labels=['Não', 'Sim'], rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(6, 4))
sns.pointplot(data=df, x='posse_de_imovel', y='renda', ci=95, markers='o', color='teal')
plt.title('Renda Média por Posse de Imóvel')
plt.xlabel('Possui Imóvel?')
plt.ylabel('Renda Média')
plt.xticks([0, 1], ['Não', 'Sim'])
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(6, 4))
sns.pointplot(data=df, x='posse_de_veiculo', y='renda', ci=95, markers='o', color='orange')
plt.title('Renda Média por Posse de Veículo')
plt.xlabel('Possui Veículo?')
plt.ylabel('Renda Média')
plt.xticks([0, 1], ['Não', 'Sim'])
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
AO final podemos observar que a posse de veículo demonstra estar mais relacionada a rendas mais altas de forma mais expressiva do que a de imóvel.
````

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('sinasc_RO_2019.csv')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
df['IDADEMAE'] = pd.to_numeric(df['IDADEMAE'], errors='coerce')

plt.figure(figsize=(8, 5))
sns.boxplot(y=df['IDADEMAE'], color='skyblue')

plt.title('Boxplot da Idade da Mãe')
plt.ylabel('Idade da Mãe')
plt.grid(True)
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
df['PESO'] = pd.to_numeric(df['PESO'], errors='coerce')

plt.figure(figsize=(8, 5))
sns.boxplot(y=df['PESO'], color='lightgreen')

plt.title('Boxplot do Peso do Bebê ao Nascer')
plt.ylabel('Peso (gramas)')
plt.grid(True)
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
df['IDADEMAE'] = pd.to_numeric(df['IDADEMAE'], errors='coerce')

plt.figure(figsize=(10, 6))
sns.histplot(df['IDADEMAE'], bins=20, kde=True, color='salmon')

plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
df['PESO'] = pd.to_numeric(df['PESO'], errors='coerce')

plt.figure(figsize=(10, 6))
sns.histplot(df['PESO'], bins=30, kde=True, color='lightblue')

plt.title('Histograma do Peso do Bebê ao Nascer')
plt.xlabel('Peso (gramas)')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```
